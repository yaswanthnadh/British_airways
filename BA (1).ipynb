```python
from scipy.stats.mstats import winsorize

df_winsor = df.copy()

# Apply Winsorization to numerical columns (capping extreme values at 5% and 95%)
for col in df.select_dtypes(include=['number']).columns:
    df_winsor[col] = winsorize(df[col], limits=[0.05, 0.05])

print("Original Data Shape:", df.shape)
print("After Winsorization (Capping):", df_winsor.shape)
```

```python
df_winsor.boxplot(figsize=(15, 10))
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
df_robust = df.copy()
num_cols = df.select_dtypes(include=['number']).columns
robust_scaler = RobustScaler()
df_robust[num_cols] = robust_scaler.fit_transform(df_robust[num_cols])
```

```python
fig, axes = plt.subplots(1, 1, figsize=(18, 6))
# Robust Scaler
sns.boxplot(data=df_robust, ax=axes)
axes.set_title("Robust Scaler (Handles Outliers)")

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
# Selecting categorical columns
cat_cols = df.select_dtypes(include=['object']).columns
```

```python
# Label Encoding
df_label_encoded = df.copy()
label_encoder = LabelEncoder()
for col in cat_cols:
    df_label_encoded[col] = label_encoder.fit_transform(df_label_encoded[col])
```

```python
# Display results
print("Label Encoded Data:")
print(df_label_encoded.head())
print("\nUnique Values in Each Column:")
print(df_label_encoded.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Define features (X) and target variable (y)
X = df_label_encoded.drop(columns=['TIER1_ELIGIBLE_PAX'])  # Features
y = df_label_encoded['TIER1_ELIGIBLE_PAX']  # Target variable
```

```python
# Perform train-test split (Stratified for imbalanced data)
# First, check the value counts of y to identify classes with only one member
class_counts = y.value_counts()
classes_to_remove = class_counts[class_counts < 2].index

# Remove rows corresponding to classes with less than 2 members
X_filtered = X[~y.isin(classes_to_remove)]
y_filtered = y[~y.isin(classes_to_remove)]

X_train, X_test, y_train, y_test = train_test_split(
    X_filtered, y_filtered, test_size=0.2, random_state=42, stratify=y_filtered
)

# Print the class distribution
print("Train set class distribution:\n", y_train.value_counts(normalize=True))
print("\nTest set class distribution:\n", y_test.value_counts(normalize=True))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Initialize the model
log_reg = LogisticRegression(random_state=42)
```

```python
from sklearn.impute import SimpleImputer

# Select numerical columns for imputation, excluding 'FLIGHT_DATE'
num_cols_for_imputation = X_train.select_dtypes(include=['number']).columns

# Median Imputation for missing values in numerical columns
imputer = SimpleImputer(strategy='median')
X_train_imputed = imputer.fit_transform(X_train[num_cols_for_imputation])
X_test_imputed = imputer.transform(X_test[num_cols_for_imputation])
```

```python
# Train the model again
log_reg.fit(X_train_imputed, y_train)
```

```python
# Make predictions on test data
y_pred = log_reg.predict(X_test_imputed)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}\n")
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Initialize Decision Tree Classifier
dt_model = DecisionTreeClassifier(random_state=42)
```

```python
# Train the model
dt_model.fit(X_train_imputed, y_train)
```

```python
# Make predictions
y_pred_dt = dt_model.predict(X_test_imputed)
```

```python
# Evaluate the model
accuracy_dt = accuracy_score(y_test, y_pred_dt)
conf_matrix_dt = confusion_matrix(y_test, y_pred_dt)
class_report_dt = classification_report(y_test, y_pred_dt)
```

```python
# Print Results
print(f"Accuracy: {accuracy_dt:.4f}")
print("\nClassification Report:\n", class_report_dt)
print("\nConfusion Matrix:\n", conf_matrix_dt)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Initialize Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
# Train the model
rf_model.fit(X_train_imputed, y_train)
```

```python
# Make predictions
y_pred_rf = rf_model.predict(X_test_imputed)
```

```python
# Evaluate the model
accuracy_rf = accuracy_score(y_test, y_pred_rf)
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
class_report_rf = classification_report(y_test, y_pred_rf)
```

```python
# Print Results
print(f"Accuracy: {accuracy_rf:.4f}")
print("\nClassification Report:\n", class_report_rf)
print("\nConfusion Matrix:\n", conf_matrix_rf)
```
